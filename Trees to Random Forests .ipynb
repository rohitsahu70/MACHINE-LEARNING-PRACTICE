```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
df
```

```python
np.random.uniform(0,1,len(df))<=0.7
```

```python
df['trainset'] = np.random.uniform(0,1,len(df))<=0.7
```

```python
df
```

```python
train = df[df['trainset']==True]
test = df[df['trainset']==False]
```

```python
features = train.columns[:4]
```

```python
pd.factorize(train['species'])
```

```python
y_train = pd.factorize(train['species'])[0]
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(train[features], y_train)
```

```python
y_pred = clf.predict(test[features])
```

```python
y_test = pd.factorize(test['species'])[0]
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred))
```

```python
iris.target_names
```

```python
predictions = iris.target_names[clf.predict(test[features])]
```

```python
print ('Predicted\tActual')
for i in range(len(predictions)):
    print (predictions[i], '\t', test.iloc[i,4])
```

# Visualizing the Random Forest

```python
clf.estimators_
```

```python
from IPython.display import Image
from sklearn.tree import export_graphviz

for i, estimator in enumerate(clf.estimators_):
    print (export_graphviz(estimator))
```

```python
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus

for i, estimator in enumerate(clf.estimators_):
    dot_data = export_graphviz(estimator)
    graph = pydotplus.graph_from_dot_data(dot_data)
    graph.write_png('tree'+str(i)+'.png')
    display(Image(graph.create_png()))
```

# Hyperparameter Tuning

```python
clf = RandomForestClassifier()
clf.get_params()
```

```python
features
```

```python
from sklearn.metrics import f1_score
clf = RandomForestClassifier(n_estimators=1)
clf.fit(train[features], y_train)
y_pred = clf.predict(test[features])
f1_score(y_test, y_pred, average='micro')
```

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [1,10,100,200,400,1000]
max_features = ['auto', 'sqrt']
max_depth = [3,6,10]
min_samples_leaf = [1,2,4,10]

random_grid = {'n_estimators': n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_leaf':min_samples_leaf
}
```

```python
clf = RandomForestClassifier()

clf_randomgrid = RandomizedSearchCV(estimator = clf, param_distributions=random_grid, cv=5, verbose=2)
clf_randomgrid.fit(train[features], y_train)
```

```python
clf_randomgrid.best_params_
```

```python
clf = RandomForestClassifier(n_estimators=10, min_samples_leaf=4, max_features='sqrt', max_depth=10)
clf.fit(train[features], y_train)
y_pred = clf.predict(test[features])
f1_score(y_test, y_pred, average='micro')
```
